```python
#import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# load dataset
df = pd.read_csv("/content/KaggleV2-May-2016.csv")
df.head()
```

```python
df.isnull().sum()
```

```python
missing = df.isnull().sum()
missing = missing[missing > 0]

if not missing.empty:
    missing.plot(kind='bar')
    plt.title("Missing Values per Column")
    plt.ylabel("Count")
    plt.show()
else:
    print("No missing values found in the dataset.")
```

The data set does not contain missing values , hence no imputattion was required.
visualization confirms data completeness

The Medical Appointment No Shows dataset was analyzed for missing values. No missing data was found in any column. Therefore, no imputation or row/column removal was required. Validation checks confirm that the dataset is complete and ready for further analysis or modeling
